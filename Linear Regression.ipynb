Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

Generating random dataset with some non-linear relationship for more realistic data

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.sin(5 * X) + np.random.randn(100, 1)  # Adding non-linearity with sin(5 * X)
```

Fit Linear Regression model using scikit-learn

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)  # Train the model on the data
```

Predict the target values

```python
y_pred = lin_reg.predict(X)
```

Model parameters (intercept and slope)

```python
print("Intercept:", lin_reg.intercept_)
print("Slope:", lin_reg.coef_)
```

Visualizing the data and the regression line

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')  # Regression line
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

Essential Metrics Calculation

Residual Sum of Squares (RSS)

RSS is the sum of the squared residuals (errors), i.e., the squared differences between the actual and predicted values.

Formula: RSS = Σ (y_actual - y_predicted)^2

```python
residuals = y - y_pred
RSS = np.sum(residuals ** 2)
print("Residual Sum of Squares (RSS):", RSS)
```

R-squared (R2) - already available through sklearn's r2_score

Formula: R² = 1 - (RSS / TSS)
TSS measures the total variance in the dependent variable.
 TSS = Σ (y_actual - y_mean)^2

```python
R2 = r2_score(y, y_pred)
print("R-squared (R2):", R2)
```

Mean Squared Error (MSE)
MSE is the average of the squared differences between the predicted and actual values.

Formula: MSE = (1/n) Σ (y_actual - y_predicted)^2

```python
MSE = mean_squared_error(y, y_pred)
print("Mean Squared Error (MSE):", MSE)
```

Mean Absolute Error (MAE):
MAE is the average of the absolute differences between the predicted and actual values.

Formula: MAE = (1/n) Σ |y_actual - y_predicted|

```python
MAE = mean_absolute_error(y, y_pred)
print("Mean Absolute Error (MAE):", MAE)
```

Prediction with new test example

```python
new_test_example = np.array([[4.5]])  # Example value
new_test_pred = lin_reg.predict(new_test_example)
print(f"Prediction for new test example (X=4.5): {new_test_pred[0][0]}")
```
